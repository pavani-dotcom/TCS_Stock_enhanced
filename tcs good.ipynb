```python
# Step 2: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import os
from google.colab import files, drive

from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import shap
import warnings
warnings.filterwarnings('ignore')
```

```python
# Step 3: Mount Google Drive

SAVE_DIR = "/content/drive/MyDrive/TCS_Stock_Project_Enhanced"
os.makedirs(SAVE_DIR, exist_ok=True)
print(f"📁 All outputs will be saved to: {SAVE_DIR}")
```

```python

# Load data
history_df = pd.read_csv('/content/drive/MyDrive/TCS_Stock_Project_Enhanced/TCS_stock_history.csv')
actions_df = pd.read_csv('/content/drive/MyDrive/TCS_Stock_Project_Enhanced/TCS_stock_action.csv')  # Assuming it has Dividends/Splits
info_df    = pd.read_csv('/content/drive/MyDrive/TCS_Stock_Project_Enhanced/TCS_stock_info.csv', header=None, names=['Key', 'Value'])

# Convert Date
history_df['Date'] = pd.to_datetime(history_df['Date'])
history_df = history_df.sort_values('Date').reset_index(drop=True)
history_df.set_index('Date', inplace=True)
history_df.sort_index(inplace=True)
```

```python
# Step 6: Feature Engineering
history_df['Log_Return'] = np.log(history_df['Close'] / history_df['Close'].shift(1))
history_df['MA_7'] = history_df['Close'].rolling(window=7).mean()
history_df['MA_30'] = history_df['Close'].rolling(window=30).mean()
history_df['Vol_7'] = history_df['Log_Return'].rolling(window=7).std() * np.sqrt(7)
history_df['Vol_30'] = history_df['Log_Return'].rolling(window=30).std() * np.sqrt(30)
history_df['Price_Change'] = history_df['Close'].diff()
history_df['High_Low_Pct'] = (history_df['High'] - history_df['Low']) / history_df['Close']
history_df['OHLC_Avg'] = (history_df[['Open','High','Low','Close']]).mean(axis=1)

for lag in [1, 2, 3, 7]:
    history_df[f'Close_Lag_{lag}'] = history_df['Close'].shift(lag)

history_df.dropna(inplace=True)

features = ['Open', 'High', 'Low', 'Volume', 'MA_7', 'MA_30', 'Vol_7', 'Vol_30',
            'Price_Change', 'High_Low_Pct', 'OHLC_Avg'] + [f'Close_Lag_{i}' for i in [1,2,3,7]]
X = history_df[features]
y = history_df['Close']

print(f"✅ Feature engineering complete. {len(features)} features created.")
```

```python
# Step 7: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
# Scale data
scaler_xgb = StandardScaler()
X_train_scaled_xgb = scaler_xgb.fit_transform(X_train)
X_test_scaled_xgb = scaler_xgb.transform(X_test)

scaler_rf = StandardScaler()
X_train_scaled_rf = scaler_rf.fit_transform(X_train)
X_test_scaled_rf = scaler_rf.transform(X_test)

print("✅ Data scaling complete.")
```

```python
# --- XGBoost ---
xgb_param_dist = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
    'gamma': [0, 0.1, 0.2]
}
xgb = XGBRegressor(random_state=42)
xgb_random = RandomizedSearchCV(xgb, xgb_param_dist, n_iter=20, cv=3,
                                scoring='neg_mean_absolute_error', n_jobs=-1, random_state=42)
xgb_random.fit(X_train_scaled_xgb, y_train)
xgb_model = xgb_random.best_estimator_

# Save XGBoost
joblib.dump(xgb_model, f'{SAVE_DIR}/tcs_xgb_model.pkl')
joblib.dump(scaler_xgb, f'{SAVE_DIR}/scaler_xgb.pkl')
print("✅ XGBoost model and scaler saved.")
```

```python
# --- Random Forest ---
rf_param_dist = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['auto', 'sqrt']
}
rf = RandomForestRegressor(random_state=42)
rf_random = RandomizedSearchCV(rf, rf_param_dist, n_iter=10, cv=3,
                               scoring='neg_mean_absolute_error', n_jobs=-1, random_state=42)
rf_random.fit(X_train_scaled_rf, y_train)
rf_model = rf_random.best_estimator_

# Save RF
joblib.dump(rf_model, f'{SAVE_DIR}/tcs_rf_model.pkl')
joblib.dump(scaler_rf, f'{SAVE_DIR}/scaler_rf.pkl')
print("✅ Random Forest model and scaler saved.")
```

```python
# --- LSTM ---
scaler_lstm = MinMaxScaler()
scaled_close = scaler_lstm.fit_transform(history_df[['Close']])

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

SEQ_LENGTH = 60
X_seq, y_seq = create_sequences(scaled_close, SEQ_LENGTH)
split = int(0.8 * len(X_seq))
X_train_lstm, X_test_lstm = X_seq[:split], X_seq[split:]
y_train_lstm, y_test_lstm = y_seq[:split], y_seq[split:]
# Build LSTM
lstm_model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(SEQ_LENGTH, 1)),
    Dropout(0.3),
    LSTM(50),
    Dropout(0.3),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(X_train_lstm, y_train_lstm, epochs=20, batch_size=32, verbose=0)

# Save LSTM model (HDF5 format)
lstm_model.save(f'{SAVE_DIR}/tcs_lstm_model.h5')
joblib.dump(scaler_lstm, f'{SAVE_DIR}/scaler_lstm.pkl')
print("✅ LSTM model and scaler saved.")
```

```python
# Step 9: Predictions
y_pred_xgb = xgb_model.predict(X_test_scaled_xgb)
y_pred_rf = rf_model.predict(X_test_scaled_rf)
lstm_pred_scaled = lstm_model.predict(X_test_lstm)
lstm_pred = scaler_lstm.inverse_transform(lstm_pred_scaled)
y_test_actual_lstm = scaler_lstm.inverse_transform(y_test_lstm.reshape(-1, 1))
# Save predictions
results_df = pd.DataFrame({
    'Date': y_test.index,
    'Actual_Close': y_test.values,
    'XGB_Pred': y_pred_xgb,
    'RF_Pred': y_pred_rf
})
results_df.to_csv(f'{SAVE_DIR}/predictions_tabular.csv', index=False)

lstm_results = pd.DataFrame({
    'Date': history_df.index[-len(y_test_actual_lstm):],
    'LSTM_Actual': y_test_actual_lstm.flatten(),
    'LSTM_Pred': lstm_pred.flatten()
})
lstm_results.to_csv(f'{SAVE_DIR}/predictions_lstm.csv', index=False)
print("✅ Prediction results saved.")
```

```python
# Step 10: Model Evaluation & Save Report
results = {
    "Model": ["XGBoost", "Random Forest", "LSTM"],
    "MAE": [
        mean_absolute_error(y_test, y_pred_xgb),
        mean_absolute_error(y_test, y_pred_rf),
        mean_absolute_error(y_test_actual_lstm, lstm_pred)
    ],
    "MSE": [
        mean_squared_error(y_test, y_pred_xgb),
        mean_squared_error(y_test, y_pred_rf),
        mean_squared_error(y_test_actual_lstm, lstm_pred)
    ],
    "R2": [
        r2_score(y_test, y_pred_xgb),
        r2_score(y_test, y_pred_rf),
        r2_score(y_test_actual_lstm, lstm_pred)
    ]
}
results_df_report = pd.DataFrame(results)
results_df_report.to_csv(f'{SAVE_DIR}/model_comparison.csv', index=False)
print("✅ Model comparison report saved.")

# Print
print("\n📊 Model Performance:")
print(results_df_report.to_string(index=False))
```

```python
# Step 11: Save Plots
plt.figure(figsize=(16, 6))

# Plot 1: Actual vs Predicted
plt.subplot(1, 3, 1)
plt.plot(y_test.values, label='Actual', color='blue')
plt.plot(y_pred_xgb, label='XGB Pred', color='orange', alpha=0.7)
plt.title('XGBoost: Actual vs Predicted')
plt.legend()
plt.grid(True)

# Plot 2: Feature Importance (XGBoost)
plt.subplot(1, 3, 2)
importance = xgb_model.feature_importances_
indices = np.argsort(importance)[::-1][:10]
plt.barh(range(10), importance[indices])
plt.yticks(range(10), [features[i] for i in indices])
plt.title('Top 10 Feature Importance (XGBoost)')
plt.xlabel('Importance')

# Plot 3: LSTM Predictions
plt.subplot(1, 3, 3)
plt.plot(y_test_actual_lstm, label='LSTM Actual', color='green')
plt.plot(lstm_pred, label='LSTM Pred', color='red', alpha=0.7)
plt.title('LSTM: Actual vs Predicted')
plt.legend()
plt.grid(True)

plt.tight_layout()
plot_path = f'{SAVE_DIR}/model_plots.png'
plt.savefig(plot_path, dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plots saved.")
```

```python
# Step 12: SHAP for XGBoost
explainer = shap.TreeExplainer(xgb_model)
shap_values = explainer.shap_values(X_test)

# Save SHAP values
np.save(f'{SAVE_DIR}/shap_values.npy', shap_values)
np.save(f'{SAVE_DIR}/shap_expected_value.npy', explainer.expected_value)

# Plot SHAP
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_test, plot_type="bar", show=False)
plt.title("SHAP Feature Importance")
plt.savefig(f'{SAVE_DIR}/shap_summary.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ SHAP analysis saved.")
```
